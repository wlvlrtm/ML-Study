```python
import matplotlib.pyplot as plt

### Bream + Smelt
plt.scatter(bream_length, bream_weight) ## Create Graph; Bream
plt.scatter(smelt_length, smelt_weight) ## Create Graph; Smelt

plt.xlabel("Length")    ## X Axis; Length
plt.ylabel("Weight")    ## Y Axis; Weight
```

```python
length = (bream_length + smelt_length)  ## Length; Bream + Smelt
weight = (bream_weight + smelt_weight)  ## Weight; Bream + Smelt

fish_data = [[len, wei] for len, wei in zip(length, weight)]    ## Total Length & Weight Data; 2D Graph
```

```python
fish_target = [1] * 35 + [0] * 14   ## Fish Answer Book
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier() ## Class Obj Create; K-NN

kn.fit(fish_data, fish_target)      ## Model Training
kn.score(fish_data, fish_target)    ## Model Evaluation
```

```python
plt.scatter(bream_length, bream_weight) ## Bream Graph; Blue
plt.scatter(smelt_length, smelt_weight) ## Smelt Graph; Orange
plt.scatter(30, 600, marker = '^')      ## New Data; Green

plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()
```

```python
print(kn.predict([[30, 600]])) ## New Data Evaluation; (1 == bream, 2 == smelt)
```
